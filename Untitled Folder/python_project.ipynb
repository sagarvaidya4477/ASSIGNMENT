```python
import requests
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Constants
API_KEY = "2fa18fe6ac3c2a956bfca974ea222f40"  # Your OpenWeatherMap API key
LOCATION = 'pune'  # Default location for fetching weather data
DB_NAME = 'weather_data.db'  # SQLite database name

def fetch_weather_data(api_key, location):
    """
    Fetch weather data from the OpenWeatherMap API.
    
    Parameters:
    - api_key (str): API key for authentication
    - location (str): Location name to fetch weather data for

    Returns:
    - dict: JSON response from the API containing weather data
    """
    url = f"https://api.openweathermap.org/data/2.5/weather?q={location}&appid={api_key}&units=metric"
    response = requests.get(url)
    response.raise_for_status()  # Raise an HTTPError if the HTTP request returned an unsuccessful status code
    return response.json()

def create_db():
    """
    Create SQLite database and table if not exists.
    """
    conn = sqlite3.connect(DB_NAME)
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS weather
                 (id INTEGER PRIMARY KEY, 
                  temperature REAL, 
                  humidity REAL, 
                  wind_speed REAL, 
                  timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
                  weather TEXT)''')
    conn.commit()
    conn.close()

def insert_weather_data(temperature, humidity, wind_speed, weather):
    """
    Insert weather data into SQLite database.
    
    Parameters:
    - temperature (float): Temperature in Celsius
    - humidity (float): Humidity in percentage
    - wind_speed (float): Wind speed in meters per second
    - weather (str): Weather condition description
    """
    conn = sqlite3.connect(DB_NAME)
    c = conn.cursor()
    c.execute('''INSERT INTO weather (temperature, humidity, wind_speed, weather) 
                 VALUES (?, ?, ?, ?)''',
              (temperature, humidity, wind_speed, weather))
    conn.commit()
    conn.close()

def import_and_clean_csv(file_path):
    """
    Import and clean CSV data.
    
    Parameters:
    - file_path (str): Path to the CSV file

    Returns:
    - DataFrame: Cleaned pandas DataFrame
    """
    df = pd.read_csv(file_path)
    df = df.rename(columns={
        'temp': 'temperature',
        'humidity': 'humidity',
        'speed': 'wind_speed',
        'dt': 'timestamp',
        'name': 'city'
    })
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    return df

def insert_csv_data(df):
    """
    Insert data from CSV file into SQLite database.
    
    Parameters:
    - df (DataFrame): pandas DataFrame containing weather data
    """
    conn = sqlite3.connect(DB_NAME)
    df.to_sql('weather', conn, if_exists='append', index=False)
    conn.commit()
    conn.close()

def plot_separate_metrics(df, city):
    """
    Plot temperature, humidity, wind speed, and a pie chart for weather distribution.
    
    Parameters:
    - df (DataFrame): pandas DataFrame containing weather data
    - city (str): City name for the plot title
    """
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['hour'] = df['timestamp'].dt.hour
    
    plt.figure(figsize=(20, 16))
    
    # Temperature Plot
    plt.subplot(2, 2, 1)
    plt.plot(df['hour'], df['temperature'], marker='o', linestyle='-', color='#FF5733', alpha=0.8)
    plt.xlabel('Hour of the Day', fontsize=14, fontweight='bold')
    plt.ylabel('Temperature (°C)', fontsize=14, fontweight='bold')
    plt.title('Temperature Over Time', fontsize=16, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    
    # Humidity Plot
    plt.subplot(2, 2, 2)
    plt.plot(df['hour'], df['humidity'], marker='x', linestyle='--', color='#33C1FF', alpha=0.8)
    plt.xlabel('Hour of the Day', fontsize=14, fontweight='bold')
    plt.ylabel('Humidity (%)', fontsize=14, fontweight='bold')
    plt.title('Humidity Over Time', fontsize=16, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Wind Speed Plot
    plt.subplot(2, 2, 3)
    plt.plot(df['hour'], df['wind_speed'], marker='s', linestyle='-.', color='#66FF66', alpha=0.8)
    plt.xlabel('Hour of the Day', fontsize=14, fontweight='bold')
    plt.ylabel('Wind Speed (m/s)', fontsize=14, fontweight='bold')
    plt.title('Wind Speed Over Time', fontsize=16, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Pie Chart
    plt.subplot(2, 2, 4)
    weather_counts = df['weather'].value_counts()
    wedges, texts, autotexts = plt.pie(
        weather_counts, 
        labels=weather_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=plt.get_cmap('Set3').colors,
        shadow=True,
        explode=[0.1]*len(weather_counts))  # Explode all slices
    plt.title('Weather Condition Distribution', fontsize=16, fontweight='bold')
    for text in autotexts:
        text.set_color('white')
        text.set_fontsize(12)
    plt.setp(texts, size=12, weight='bold')

    # Add City Name to Bottom
    plt.figtext(0.5, 0.01, f"City: {city}", ha='center', fontsize=18, fontweight='bold', color='black')

    plt.tight_layout()
    plt.show()

def main(city):
    """
    Main function to execute the program steps.
    
    Parameters:
    - city (str): City name to filter data for visualization
    """
    # Uncomment the lines below to enable database creation and data fetching
    # create_db()
    # data = fetch_weather_data(API_KEY, LOCATION)
    # temperature = data['main']['temp']
    # humidity = data['main']['humidity']
    # wind_speed = data['wind']['speed']
    # weather = data['weather'][0]['main']
    # insert_weather_data(temperature, humidity, wind_speed, weather)
    
    # Load and filter data
    df = import_and_clean_csv("final_df.csv")
    df = df[df['city'] == city]
    
    # Plot data
    plot_separate_metrics(df, city)

if __name__ == "__main__":
    main('Mumbai')
```
